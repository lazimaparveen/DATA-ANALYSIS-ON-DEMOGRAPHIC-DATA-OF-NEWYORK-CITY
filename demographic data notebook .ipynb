```python
ndf.groupby("School_Name").Total_Enrollment.sum().nlargest(5)
```

```python
ndf.info()
```

```python
ndf.nlargest(5,"Total_Enrollment")
```

```python
plt.figure(figsize = (12,8))
ndf.groupby("School_Name").Total_Enrollment.sum().nlargest(5).plot(kind='bar', fontsize = 15)
plt.title("Most Popular Schools",fontsize = 20)
plt.ylabel("Total Enrollment", fontsize = 15)
plt.show()
```

```python
plt.figure(figsize = (12,8))
ndf.groupby("Category").Total_Enrollment.sum().nlargest(5).plot(kind='bar', fontsize = 15)
plt.title("Visualisation Based On Category",fontsize = 20)
plt.ylabel("Total Enrollment", fontsize = 15)
plt.show()
```
